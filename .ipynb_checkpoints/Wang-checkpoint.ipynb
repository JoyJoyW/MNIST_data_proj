```python
from scipy.io import loadmat
import matplotlib.pyplot as plt

#Loads in dataset
# MNIST dataset is a widely used dataset for handwritten digit recognition
# Contains 70k images each representing a number 0-9
mnist = loadmat(r"mnist-original.mat")
mnist_data = mnist["data"].T
mnist_label = mnist["label"][0]
```

```python
#normalizes data
mnist_data_normalized = mnist_data / 255.0
```

```python
# Computes and visualizes the average image of each digit from 0 to 9
for i in range(10):
    plt.figure(figsize=(8, 6))
    avg = sum(mnist_data_normalized[mnist_label==i])/len(mnist_label[mnist_label==i])
    plt.imshow(avg.reshape(28,28))
    
    # Labels and title
    plt.title("Visualization of Sample Image")
    
    # Show plot
    plt.show()
```

## Trains a linear SVM classifier on the training dataset, and evaluate the model on the test data to see how it performs in distinguishing between digits 4 and 9.

```python
from sklearn import svm
clf = svm.SVC(kernel = "linear")
X = mnist_data_normalized[(mnist_label==4) | (mnist_label==9)]
y = mnist_label[(mnist_label==4) | (mnist_label==9)]
y[y==4] = 0
y[y==9] = 1
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=12000, random_state=42, stratify=y)
X_train, X_test, y_train, y_test
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
# Reports precision, accuracy, F1 score, and recall for model

from sklearn.metrics import precision_score, accuracy_score, recall_score, f1_score

precision = precision_score(y_test, y_pred)
precision
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
recall = recall_score(y_test, y_pred)
recall
```

```python
f1_score = f1_score(y_test, y_pred)
f1_score
```

```python
# Implements a rbf kernel support vector machine
rbf_svm = svm.SVC(kernel='rbf')
rbf_svm.fit(X_train, y_train)
```

```python
y_pred_rbf = rbf_svm.predict(X_test)
```

```python
precision_rbf = precision_score(y_test, y_pred_rbf)
precision_rbf
```

```python
accuracy_rbf = accuracy_score(y_test, y_pred_rbf)
accuracy_rbf
```

```python
recall_rbf = recall_score(y_test, y_pred_rbf)
recall_rbf
```

```python
from sklearn.metrics import f1_score

f1_score_rbf = f1_score(y_test, y_pred_rbf)
f1_score_rbf
```

```python
from sklearn.metrics import classification_report
# Classification report for RBF kernel
print(classification_report(y_test, y_pred_rbf))
```

```python
# Implement a multi-class SVM classifier
svm_multi = svm.SVC(kernel = 'rbf', decision_function_shape='ovr')
X_multi = mnist_data_normalized
y_multi = mnist_label
```

```python
X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(X_multi, y_multi, train_size=60000, random_state=42, stratify=y_multi)
X_train_multi, X_test_multi, y_train_multi, y_test_multi
```

```python
svm_multi.fit(X_train_multi, y_train_multi)
```

```python
y_pred_multi = svm_multi.predict(X_test_multi)
```

```python
# Reports macro/micro precision, macro/micro recall, macro/micro F1 score,
# and accuracy for multi-class SVM.
micro_precision = precision_score(y_test_multi, y_pred_multi, average='micro')
micro_recall = recall_score(y_test_multi, y_pred_multi, average='micro')
micro_f1 = f1_score(y_test_multi, y_pred_multi, average='micro')
micro_precision, micro_recall, micro_f1
```

```python
macro_precision = precision_score(y_test_multi, y_pred_multi, average='macro')
macro_recall = recall_score(y_test_multi, y_pred_multi, average='macro')
macro_f1 = f1_score(y_test_multi, y_pred_multi, average='macro')
macro_precision, macro_recall, macro_f1
```

```python
accuracy_multi = accuracy_score(y_test_multi, y_pred_multi)
accuracy_multi
```
